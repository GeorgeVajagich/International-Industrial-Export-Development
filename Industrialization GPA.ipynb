```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
Years=[]
for i in range (1962,2012):
    i=str(i)
    Years.append(i)
print(Years)

AgNatList=[]
LightList=[]
HeavyList=[]
GPAList=[]
```

```python
for y in Years:
    # importing csv
    data=pd.read_csv(r"C:\Users\Geord\Desktop\Export_Data\South Korea\South Korea " +y+".csv")

    # checking when category each row is
    y=(data["Category"])
    category=[]
    for str in y:
        category.append(str)

    # checking when how much trade is in each cow
    y=(data["Trade Value"])
    value=[]
    for str in y:
        value.append(str)

    # creating variables to keep track of how much is exported of the following macro groups    
    # Agriculture/Natural Resources
    # Light Indsutry
    # Heavy Industry
    AgNat=0
    Light=0
    Heavy=0

    # lists of what categories fall into what macro groups
    AgNatCats=['Beer, Spirits and cigarrettes','Food Processing','Petrochemicals','Inorganic salts and acids','Coal','Mining','Oil','Precious Stones','Processed minerals','Cereals and vegetable oils','Tropical treecrops and flowers','Tobacco','Fruit','Misc Agriculture','Fish & Seafood','Meat and eggs','Animal Fibers','Milk & cheese','Leather']
    LightCats=['Construction materials and equipment','Pulp and paper','Textile & Fabrics','Garments']
    HeavyCats=['Machinery','Electronics','Aircraft','Boilers','Ships','Metal products','Home and office products','Other Chemicals','Agrochemicals','Chemicals and health related products']
    # 'Not classified' and 'Unspecified' are categories not included in any of the macro groups

    # calculating how much is exported of each macro category
    for int in range (0,len(category)):
        c=category[int]
        if c in AgNatCats:
            AgNat=AgNat+value[int]
        if c in LightCats:
            Light=Light+value[int]
        if c in HeavyCats:
            Heavy=Heavy+value[int]
        
    pAgNat=(AgNat/(AgNat+Light+Heavy))
    pLight=(Light/(AgNat+Light+Heavy))
    pHeavy=(Heavy/(AgNat+Light+Heavy))
    GPA=(pLight*2+pHeavy*4)

    AgNatList.append(pAgNat)
    LightList.append(pLight)
    HeavyList.append(pHeavy)
    GPAList.append(GPA)
```

```python
plt.axis([0,50,0,1])
plt.plot(Years,AgNatList,label="Agriculture/Natural Resources",c="yellow")
plt.plot(Years,LightList,label="Light Indsutry",c="green")
plt.plot(Years,HeavyList,label="Heavy Industry",c="blue")
plt.xticks(range(0, 50,4))
plt.title("South Korea",size=20)
plt.legend()
```
